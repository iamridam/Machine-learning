```python
import pandas as pd
```

```python
ndna_file = 'final_amino_acid_result_ndna.csv'
```

```python
dna_file = 'final_amino_acid_result_dna.csv'
```

```python
dna = pd.read_csv(dna_file,skiprows= lambda x: True if x%2 else False)
```

```python
ndna = pd.read_csv(ndna_file,skiprows= lambda x: True if x%2 else False)
```

```python
dna.shape
```

```python
ndna.shape
```

```python
label = pd.DataFrame([1 for i in range(1299)]+[0 for i in range(1750)])
```

```python
data = pd.concat([dna,ndna])
```

```python
data.head()
```

```python
label.T
```

```python
from keras.utils import to_categorical
label_binary = to_categorical(label)
```

```python
data.index = data.ID
```

```python
data = data.drop('ID',axis=1)
```

```python
data.head()
```

```python
#Dependencies
import keras
from keras.models import Sequential
from keras.layers import Dense

# Neural network
model = Sequential()
model.add(Dense(512, input_dim=20, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(data, label, epochs=30)
```
